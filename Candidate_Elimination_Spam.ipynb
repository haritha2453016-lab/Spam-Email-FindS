```python
import pandas as pd

# Creating the Spam Email Dataset
data = {
    'Contains_Link': ['Yes', 'Yes', 'No', 'Yes', 'No'],
    'Contains_Offer': ['Yes', 'Yes', 'Yes', 'No', 'No'],
    'Contains_Attachment': ['No', 'No', 'No', 'No', 'Yes'],
    'Sender_Known': ['No', 'Yes', 'No', 'No', 'Yes'],
    'Spam': ['Yes', 'Yes', 'Yes', 'No', 'No']
}

df = pd.DataFrame(data)
df
```

```python
attributes = df.iloc[:, :-1].values
target = df.iloc[:, -1].values

attributes, target
```

```python
def candidate_elimination(concepts, target):
    # Initialize S and G
    S = list(concepts[0])
    G = [['?' for _ in range(len(S))]]

    for i in range(len(concepts)):
        if target[i] == 'Yes':  # Positive example
            for j in range(len(S)):
                if concepts[i][j] != S[j]:
                    S[j] = '?'
            G = [g for g in G if all(
                g[j] == '?' or g[j] == S[j] for j in range(len(S))
            )]

        else:  # Negative example
            new_G = []
            for g in G:
                for j in range(len(S)):
                    if S[j] != '?' and concepts[i][j] != S[j]:
                        new_hypothesis = g.copy()
                        new_hypothesis[j] = S[j]
                        new_G.append(new_hypothesis)
            G = new_G

    return S, G
```

```python
S_final, G_final = candidate_elimination(attributes, target)

print("Final Specific Hypothesis (S):")
print(S_final)

print("\nFinal General Hypotheses (G):")
for g in G_final:
    print(g)
```
