```python
import pandas as pd

# Creating the Spam Email Dataset
data = {
    'Contains_Link': ['Yes', 'Yes', 'No', 'Yes', 'No'],
    'Contains_Offer': ['Yes', 'Yes', 'Yes', 'No', 'No'],
    'Contains_Attachment': ['No', 'No', 'No', 'No', 'Yes'],
    'Sender_Known': ['No', 'Yes', 'No', 'No', 'Yes'],
    'Spam': ['Yes', 'Yes', 'Yes', 'No', 'No']
}

df = pd.DataFrame(data)
df
```

```python
# Splitting attributes and target
attributes = df.iloc[:, :-1]
target = df.iloc[:, -1]

attributes, target
```

```python
# Find-S Algorithm Implementation

def find_s_algorithm(attributes, target):
    hypothesis = None

    for i in range(len(target)):
        if target[i] == 'Yes':   # Positive example
            if hypothesis is None:
                hypothesis = list(attributes.iloc[i])
            else:
                for j in range(len(hypothesis)):
                    if hypothesis[j] != attributes.iloc[i, j]:
                        hypothesis[j] = '?'
    return hypothesis
```

```python
final_hypothesis = find_s_algorithm(attributes, target)

print("Final Hypothesis:")
print(final_hypothesis)

```
